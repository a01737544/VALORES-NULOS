SUSTITUCIÓN DE VALORES NULOS

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
df["review_scores_accuracy"]=df["review_scores_accuracy"].fillna(round(df["review_scores_accuracy"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```


Decidí reemplazar los valores nulos con la media, porque esta variable tiene una distribución relativamente estable y el promedio me ayuda a mantener la tendencia central sin distorsionar los datos.

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
df["review_scores_cleanliness"]=df["review_scores_cleanliness"].fillna(round(df["review_scores_cleanliness"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

Aquí usé la mediana, ya que esta medida no se ve afectada por valores atípicos y me permite imputar los nulos de una forma más robusta.

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["host_listings_count"] = df["host_listings_count"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

A esta variable le apliqué forward fill (ffill), porque tiene registros que suelen repetirse y de esa manera aproveché la continuidad de los datos sin inventar valores.

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["host_total_listings_count"] = df["host_total_listings_count"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

También usé ffill, ya que la lógica es parecida: los valores anteriores me ayudan a rellenar sin perder coherencia en los datos.

```python
#Sexto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
df["host_has_profile_pic"]=df["host_has_profile_pic"].fillna(method="bfill") 
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

Usé backfill (bfill) porque es una variable categórica con valores repetitivos y rellenar con el valor siguiente me pareció adecuado.

```python
#Sexto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
df["host_response_time"]=df["host_response_time"].fillna(method="bfill") 
```
