```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('Ventas_totales.csv')
data.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=data.isnull().sum().sum()
valores_nulos
```

```python
#Valores nulos por filas
data[data.isnull().any(axis=1)]
```

SUSTITUCIÓN DE VALORES NULOS

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data["subtotal_ventas_alimentos_bebidas"]=data["subtotal_ventas_alimentos_bebidas"].fillna(round(data["subtotal_ventas_alimentos_bebidas"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data['subtotal_ventas_alimentos_bebidas'].isnull().sum()
valores_nulos
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
data["tarjetas_credito"]=data["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=data['tarjetas_credito'].isnull().sum()
valores_nulos
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto 
data["verduleria_fruteria"]= data["verduleria_fruteria"].fillna(3508831.994) 
data
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data["almacen"] =data["almacen"].fillna(method="ffill") 
data
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Sexto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data["salon_ventas"] =data["salon_ventas"].fillna(method="bfill") 
data
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
data["otros"] =data["otros"].fillna(method="bfill") 
data
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data[["tarjetas_debito", "otros_medios", "bebidas","panaderia", "lacteos", "carnes", "alimentos_preparados_rotiseria", "indumentaria_calzado_textiles_hogar","electronicos_articulos_hogar"]] = data[["tarjetas_debito", "otros_medios", "bebidas","panaderia", "lacteos", "carnes", "alimentos_preparados_rotiseria", "indumentaria_calzado_textiles_hogar","electronicos_articulos_hogar"]].fillna(data[["tarjetas_debito", "otros_medios", "bebidas","panaderia", "lacteos", "carnes", "alimentos_preparados_rotiseria", "indumentaria_calzado_textiles_hogar","electronicos_articulos_hogar"]].mean())
data
```
