```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga de un archivo .xlsx desde un libro con varias páginas 
df= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2022', skiprows=3)
df
```

```python
#Identificar valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
#Valores nulos por filas
df[df.isnull().any(axis=1)]
```

SUSTITUCIÓN DE VALORES NULOS

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
df["Otros"]=df["Otros"].fillna(round(df["Otros"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

Se siguen obteniendo valores nulos en "Otros" ya que podrían haber valores no numericos o no hay datos, entonces procedemos a verificar el tipo de datos.


```python
# Verifica el tipo de datos
print(df["Otros"].dtype)

# Verifica la media
print(df["Otros"].mean())

# Verifica valores únicos
print(df["Otros"].unique())
```

Podemos observar que no hay datos numéricos, entonces eliinaremos la columna.

```python
# Elimina la columna "Otros" del DataFrame
df = df.drop(columns=["Otros"])
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["Poliza"] =df["Poliza"].fillna(method="ffill") 
df
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

Primero nos aseguramos de que la columna folio fuera numerica

```python
df["Folio"] = pd.to_numeric(df["Folio"], errors="coerce")
df["Folio"]
```

Despúes ya empezamos el método de sustitucion 

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
df["Folio"]=df["Folio"].fillna(round(df["Folio"].mean(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["MP"] =df["MP"].fillna(method="ffill") 
df
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
df["TC"]=df["TC"].fillna(round(df["TC"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Identificar valores nulos por dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```
