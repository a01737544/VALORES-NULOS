```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name="2020", skiprows =5)
df
```

```python
#IDENTIFICAR VALORES NULOS POR COLLUMNA

valores_nulos = df.isnull().sum()
valores_nulos
```

```python
#IDENTIFICAR VALORES NULOS EN TOTAL
valores_nulos2 = df.isnull().sum().sum()
valores_nulos2
```

```python
#VALORES NULOS POR FILA
df[df.isnull().any(axis=1)]
```

```python
#CONOCER EL PORCENTAJE DE MISSING VALUES POR COLUMNA
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(8)
```

```python
df[["POLIZA", "GASTO", "TC", "IVA", "FOLIO", "IMPORTE", "TIPO"]].dtypes
```

```python
#POLIZA TIENE UN PORCENTAJE DE MISSING VALUES DEL 99%, EN UN ANÁLISIS REAL NO SERÍA UNA VARIABLE RELEVANTE POR LA CANTIDAD DE
#INFORMACIÓN DESCONOCIDA

#AL UNA VARIABLE CATEGÓRICA NO PUEDO USAR MEDIA, POR TANTO OCUPO LA MODA. SIN EMBARGO, AL SER PÓLIZAS, EL ANÑALISIS
#INDICA QUE EL ÚLTIMO DATO REPRESENTA TAMBIÉN LOS DE LAS FILAS SUPERIORES

#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
df["POLIZA"] =df["POLIZA"].fillna(method="bfill")
df
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(8)
```

```python
#DEBIDO A QUE NO SE TERMINA DE RELLENAR LA INFORMACIÓN, APLICARÉ AHORA UN RELLENO MANUEAL STRING

#Sustituir valores nulos por un valor numérico en  concreto
df["POLIZA"]= df["POLIZA"].fillna("SIN REGISTRO")
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(8)
```

```python
#HAY MAYOR CONCENTRACIÓN DE DATOS EN LA PARTE FINAL, PERO AL SER 74% DE
#DATOS NULOS HARÉ UN RELLENO MANUEAL STRING DE SIN REGISTRO

#Sustituir valores nulos por un valor numérico en  concreto
df["GASTO"]= df["GASTO"].fillna("SIN REGISTRO")
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(8)
```

```python
#PARA TC, DECIDÍ OCUPAR LA MEDIANA YA QUE SOLO HAY 9 TIPOS DE DATOS EN LA COLUMNA,
# Y EN SU MAYORÍA TOMAN EL VALOR DE 1.

#Sustituir valores nulos con mediana
df["TC"]=df["TC"].fillna(round(df["TC"].median(),1))
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(8)
```

```python
#PARA EL CASO DE IVA, RELLENARÉ CON EL VALOR DE 0, YA QUE ANALIZANDO LA BASE DE DATOS, HAY UNA RELACIÓN ENTRE LA COLUMNA DE IVA
#Y LA COLUMNA DE DESCRIPCIÓN, Y HAY ALGUNOS CASOS EN LOS QUE NO SE GENERA IVA

#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
df["IVA"]= df["IVA"].fillna(0)
df
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(5)
```

```python
#PARA FOLIO OCUPARÉ UN FORWARD FILL YA QUE SE REPITEN CON BASE A LA FECHA EN LA QUE SE REGISTRAN

#Quinto método de sustitución de valores nulos
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
df["FOLIO"] =df["FOLIO"].fillna(method="ffill")
df
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(5)
```

```python
#PARA IMPORTE ESTARÉ OCUPANDO TAMBIÉN UN RELLENADO MANUAL DE VALOR DE 0, YA QUE TAMBIÉN PRESENTA RELACIÓN CON
#LA COLUMNA DE IVA Y DESCRIPCIÓN.

#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
df["IMPORTE"]= df["IMPORTE"].fillna(0)
df
```

```python
#Corroboramos valores nulos
porcentaje_missing = (df.isna().mean() * 100).sort_values(ascending=False)
porcentaje_missing.head(5)
```

```python
#FINALMENTE, PARA LA COLUMNA DE TIPO, SOLO EXISTE UN DATO FALTANTE, Y AL RELACIONARLA CON SU TIPO DE GASTO VI QUE
#PERTENECÍA A COMUBUSTIBLE. HACIENDO ESTA RELACIÓN OBSERVÉ QUE EL MEJOR VALOR PARA ESTE DATO FALTANTE SERÍA "I"

#PARA EL CASO DE IVA, RELLENARÉ CON EL VALOR DE 0, YA QUE ANALIZANDO LA BASE DE DATOS, HAY UNA RELACIÓN ENTRE LA COLUMNA DE IVA
#Y LA COLUMNA DE DESCRIPCIÓN, Y HAY ALGUNOS CASOS EN LOS QUE NO SE GENERA IVA

#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en  concreto
df["TIPO"]= df["TIPO"].fillna("I")
df
```

```python
#COMPROBAR QUE NO EXISTAN VALORES NULOS POR COLUMNA

valores_nulos = df.isnull().sum()
valores_nulos
```
