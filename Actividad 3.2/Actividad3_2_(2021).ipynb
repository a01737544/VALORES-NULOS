```python
#Imputación de datos en Folio

import pandas as pd
import numpy as np

año2021["FOLIO"].astype(str)

# Detectar faltantes (NaN o vacío)
is_missing = año2021["FOLIO"].isna()

# Contar cuántos faltantes hay
n_missing = is_missing.sum()

# Generar IDs tipo "1IM", "2IM", ...
new_ids = [f"{i}IM" for i in range(1, n_missing + 1)]

# Asignar
año2021.loc[is_missing, "FOLIO"] = new_ids
```

```python
#Imputación de faltantes en MP
año2021['MP'] = año2021['MP'].fillna('pago')
display(año2021.isnull().sum())
```

```python
#Imputación de faltantes en póliza
año2021['POLIZA'] = año2021['POLIZA'].bfill()
display(año2021.isnull().sum())
```

```python
# Ver las ultimas 10 filas
display(año2021.tail(10))
```

```python
año2021['POLIZA'] = año2021['POLIZA'].ffill()
display(año2021.isnull().sum())
```
